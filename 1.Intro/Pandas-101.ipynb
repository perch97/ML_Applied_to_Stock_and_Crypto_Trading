```python
df_new = data.copy()
```

```python
df_new.loc[data['Close'].shift(-1)>data['Close'],'TARGET'] = 1
df_new.loc[data['Close'].shift(-1)<=data['Close'],'TARGET'] = -1
df_new.head()
```

```python
i = 0 
for index, row in df_new.iterrows():
    print(index,row['Close'],row['TARGET'])
    if i>=4:
        break
    i +=1
```

# Extract Arrays and Values from DF

```python
#Get Close prices
df_ex = df_new.copy()
close_prices =df_ex['Close'].values
list(close_prices[:4])
```

```python
#Change Close Prices
mod_close_prices = close_prices / 2
mod_close_prices[:4]
```

```python
df_ex['Mod_close'] = mod_close_prices
df_ex
```

```python
#Get just one item
price = df_ex['Close'].iloc[1:2].item()
price
```

# Multiple Adjustment

```python
#Change Close, Volume and Open
df_ex[['Open','Close','Volume']] = df_ex[['Open','Close','Volume']]/df_ex[['Open','Close','Volume']].max()
df_ex.tail()
```
